```python
import pandas as pd
```

## Reading data from URL

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
read_url = pd.read_html(url,header=[0])
df = read_url[0]
df.head(10)
```

## Renaming columns

```python
df = df.rename(index=str, columns={'Postcode':'Postal Code','Neighbourhood':'Neighborhood'})
df.head()
```

## Ignoring not assigned values

```python
df = df[df.Borough !='Not assigned']
df.reset_index(drop=True,inplace=True)
df.head()
```

## Groupby postal code

```python
df = df.groupby('Postal Code',as_index=False).agg(lambda x: ','.join(set(x.dropna())))
df.head()
```

## If a cell has a borough but a Not assigned neighborhood, then the neighborhood will be the same as the borough.

```python
df.loc[df['Neighborhood'] == 'Not assigned','Neighborhood'] = df['Borough']
df.head()
```

## Neighbourhoods in Toronto

```python
Toronto = df_merge[df_merge['Borough'].str.contains('Toronto')].reset_index(drop=True)
Toronto.head()
```

## latitude and longitude values

```python
address = 'Toronto'
geolocator = Nominatim()
location = geolocator.geocode(address)
Latitude = location.latitude
Longitude = location.longitude
print('The geograpical coordinate of Toronto are{},{}.'.format(latitude,longitude))
```

## Toronto map

```python
Toronto_map = folium.Map(location=[Latitude, Longitude], zoom_start=11)

for lat, lng, label in zip(Toronto['Latitude'], Toronto['Longitude'], Toronto['Neighborhood']):
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(Toronto_map)  
    
Toronto_map
```
